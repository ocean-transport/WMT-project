```python
ecco = xr.open_zarr(gcfs_w_token.get_mapper('pangeo-tmp/stb2145/ECCO/datasets/wmt_budget_terms.zarr'))
sose = xr.open_zarr(gcfs_w_token.get_mapper('pangeo-tmp/stb2145/SOSE/datasets/wmt_budget_terms/wmt_budget_terms.zarr'))
soda = xr.open_zarr(gcfs_w_token.get_mapper('pangeo-tmp/stb2145/SODA/datasets/wmt_budget_terms/wmt_budget_terms.zarr'))
```

```python
psi_ecco = ecco.psi_vel.drop('time_snp').sel(sigma2_bin=slice(1037.115, 1037.5)) / 1e6
psi_sose = sose.psi_vel.sel(sigma2_bin=slice(1037.115, 1037.5)) / 1e6
psi_soda = soda.psi_vel.sel(sigma2_bin=slice(1037.185, 1039))
```

```python
def clim(da):
    clim = da.groupby('time.month').mean(dim='time')
    #anom_rolled = anom.rolling(time=12).mean(dim='time')
    return clim
```

```python
clim_ecco = clim(psi_ecco)
clim_sose = clim(psi_sose)
clim_soda = clim(psi_soda)

anom_ecco = (psi_ecco.groupby('time.month') - clim_ecco).rolling(time=12).mean(dim='time')
anom_sose = (psi_sose.groupby('time.month') - clim_sose).rolling(time=73).mean(dim='time')
anom_soda = (psi_soda.groupby('time.month') - clim_soda).rolling(time=36).mean(dim='time')
```

## Mean

```python
mean_ecco = ecco.psi_vel.mean('time') / 1e6
mean_sose = sose.psi_vel.mean('time') / 1e6
mean_soda = soda.psi_vel.mean('time')
```

```python
plt.figure(figsize=(12,5))

mean_ecco.plot(color='green', label='ecco')
mean_sose.plot(color='red', label='sose')
mean_soda.plot(color='lightslategray', label='soda')

plt.ylabel(r'$\Psi$ [Sv]', fontsize=13)
plt.xlabel('Year', fontsize=13)
plt.title(r'Time Mean $\Psi$', fontsize=18)
plt.xlim(1036.1, 1037.3)
plt.legend()
plt.grid();
```

## Climatology

```python
bot_clim_ecco = - clim_ecco.isel(sigma2_bin=-1) + clim_ecco.isel(sigma2_bin=0)
bot_clim_sose = - clim_sose.isel(sigma2_bin=-1) + clim_sose.isel(sigma2_bin=0)
bot_clim_soda = - clim_soda.isel(sigma2_bin=-1) + clim_soda.isel(sigma2_bin=0)
```

```python
plt.figure(figsize=(12,5))

bot_clim_ecco.plot(color='green', label='ecco')
bot_clim_sose.plot(color='red', label='sose')
bot_clim_soda.plot(color='lightslategray', label='soda')

plt.ylabel(r'$\Psi$ [Sv]', fontsize=13)
plt.xlabel('Year', fontsize=13)
plt.title(r'Climatological $\Psi$', fontsize=18)
plt.legend()
plt.grid();
```

## Anomaly

```python
bot_ecco = - anom_ecco.isel(sigma2_bin=-1) + anom_ecco.isel(sigma2_bin=0)
bot_sose = - anom_sose.isel(sigma2_bin=-1) + anom_sose.isel(sigma2_bin=0)
bot_soda = - anom_soda.isel(sigma2_bin=-1) + anom_soda.isel(sigma2_bin=0)
```

```python
plt.figure(figsize=(12,5))

bot_ecco.plot(color='green', label='ecco')
bot_sose.plot(color='red', label='sose')
bot_soda.plot(color='lightslategray', label='soda')

plt.ylabel(r'$\Psi$ [Sv]', fontsize=13)
plt.xlabel('Year', fontsize=13)
plt.title(r'Anomalous $\Psi$', fontsize=18)
plt.legend()
plt.grid();
```
